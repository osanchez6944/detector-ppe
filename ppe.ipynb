```python
import os

def revisar_carpetas(ruta_base):
    if not os.path.exists(ruta_base):
        print(f"❌ La carpeta raíz '{ruta_base}' NO EXISTE. ¿Subiste el dataset?")
        return
    
    print(f"✅ Carpeta raíz encontrada: {ruta_base}")
    for root, dirs, files in os.walk(ruta_base):
        level = root.replace(ruta_base, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f"{indent}📂 {os.path.basename(root)}/ ({len(files)} archivos)")

# Ejecutar revisión
revisar_carpetas('/content/Dataset_PPE')
```

```python
import os

# 📁 Ruta base de tu dataset
base_path = "ppe-factory-1"

# Carpeta de entrenamiento (ajusta si es necesario)
images_path = os.path.join(base_path, "train", "images")
labels_path = os.path.join(base_path, "train", "labels")

# 📊 Listas de archivos
imagenes = [f for f in os.listdir(images_path) if f.endswith((".jpg", ".png", ".jpeg"))]
etiquetas = [f for f in os.listdir(labels_path) if f.endswith(".txt")]

# 🔢 Conteos
print(f"🖼️ Total imágenes: {len(imagenes)}")
print(f"🏷️ Total etiquetas: {len(etiquetas)}")

# 🔍 Verificación
imagenes_sin_etiqueta = []
etiquetas_sin_imagen = []

# Crear sets sin extensión
img_names = set(os.path.splitext(f)[0] for f in imagenes)
lbl_names = set(os.path.splitext(f)[0] for f in etiquetas)

# Buscar imágenes sin etiqueta
for nombre in img_names:
    if nombre not in lbl_names:
        imagenes_sin_etiqueta.append(nombre)

# Buscar etiquetas sin imagen
for nombre in lbl_names:
    if nombre not in img_names:
        etiquetas_sin_imagen.append(nombre)

# 📢 Resultados
print("\n🔎 RESULTADOS:")

if not imagenes_sin_etiqueta and not etiquetas_sin_imagen:
    print("✅ Todo está correcto: cada imagen tiene su etiqueta")
else:
    if imagenes_sin_etiqueta:
        print(f"\n⚠️ Imágenes sin etiqueta: {len(imagenes_sin_etiqueta)}")
        print(imagenes_sin_etiqueta[:10])  # muestra solo 10

    if etiquetas_sin_imagen:
        print(f"\n⚠️ Etiquetas sin imagen: {len(etiquetas_sin_imagen)}")
        print(etiquetas_sin_imagen[:10])
```
